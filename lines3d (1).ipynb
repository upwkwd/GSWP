
# Parametric Curve


This example demonstrates plotting a parametric curve in 3D.


```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.sin(1.5*theta)
x = (2 + np.cos(1.5*theta))*np.cos(theta)
y = (2 + np.cos(1.5*theta))*np.sin(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
z = np.cos(20*theta)
x = (4 + np.sin(20*theta))*np.cos(theta)
y = (4 + np.sin(20*theta))*np.sin(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```
